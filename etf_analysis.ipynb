## 여러 개의 변수로 서브플롯 그리기

### PairGrid, barplot

```python
g = sns.PairGrid(data=df, 
             x_vars="유형", 
             y_vars=["거래량", "거래대금(백만)", "시가총액(억)", "3개월수익률"],
             aspect=5)
g.map(sns.barplot)
g.set(xlabel="시장구분")
```

### subplots 이용

```python
row_cols = ["거래량", "거래대금(백만)", "시가총액(억)", "3개월수익률"]
```

```python
fig, axes = plt.subplots(nrows=len(row_cols), figsize=(10,12))
for i, r in enumerate(row_cols):
    sns.barplot(data=df, x="유형", y=r, ax=axes[i])
```

```python
fig, axes = plt.subplots(nrows=len(row_cols), figsize=(15,20))
for i, r in enumerate(row_cols):
    sns.barplot(data=df, x="브랜드", y=r, 
                ax=axes[i], ci=None).set_title(f"브랜드별 평균 {r}")
```

### 브랜드별 인버스

```python
fig, axes = plt.subplots(nrows=len(row_cols), figsize=(15,20))
for i, r in enumerate(row_cols):
    g = sns.barplot(data=df, x="브랜드", y=r, hue="인버스",
                ax=axes[i], ci=None)
    g.set_title(f"브랜드별 평균 {r}")
    g.legend(loc=1)
```

```python
df.groupby(["브랜드","인버스"])[row_cols].mean()
```

### 브랜드별 레버리지

```python
fig, axes = plt.subplots(nrows=len(row_cols), figsize=(15,20))
for i, r in enumerate(row_cols):
    g = sns.barplot(data=df, x="브랜드", y=r, hue="레버리지",
                ax=axes[i], ci=None)
    g.set_title(f"브랜드별 평균 {r}")
    g.legend(loc=1)
```

```python
gb = df.groupby(["브랜드","레버리지"])[row_cols].mean()
gb.style.background_gradient()
```
