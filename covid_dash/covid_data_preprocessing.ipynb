```python
import pandas as pd
```

```python
url_confirmed = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
url_deaths = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
url_recovered = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
```

```python
confirmed = pd.read_csv(url_confirmed)
confirmed.head()
```

```python
# 행열 변환을 통해 dataframe을 reshape
date1 = confirmed.columns[4:]
total_confirmed = confirmed.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                                 value_vars=date1, var_name='date', value_name='confirmed')
total_confirmed.tail()
```

```python
deaths = pd.read_csv(url_deaths)
```

```python
date2 = deaths.columns[4:]
total_deaths = deaths.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
                           value_vars=date2, var_name='date', value_name='death')
total_deaths.tail()
```

```python
recovered = pd.read_csv(url_recovered)
```

```python
date3 = recovered.columns[4:]
total_recovered = recovered.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
                                 value_vars=date3, var_name='date', value_name='recovered')
total_recovered[total_recovered['recovered'] != 0].sort_values(by='recovered')
```

```python
# 재구조화 시킨 3가지 데이터를 병합
covid_data = total_confirmed.merge(right = total_deaths, how = 'left',
                                   on = ['Province/State', 'Country/Region', 'date', 'Lat', 'Long'])
covid_data.head()
```

```python
covid_data = covid_data.merge(right = total_recovered, how = 'left',
                              on = ['Province/State', 'Country/Region', 'date', 'Lat', 'Long'])
covid_data.head()
```

```python
# date 컬럼 type을 datetime으로 변환
covid_data['date'] = pd.to_datetime(covid_data['date'])
```

```python
# 결측치 확인
covid_data.isna().sum()
```

```python
# recovered 결측치를 0으로 치환
covid_data['recovered'] = covid_data['recovered'].fillna(0)
```

```python
covid_data.isna().sum()
```

```python
# active 환자 컬럼 추가
covid_data['active'] = covid_data['confirmed'] - covid_data['death'] - covid_data['recovered']
```

```python
value_confirmed = covid_data_2[covid_data_2['Country/Region'] == 'Vietnam']['confirmed'].iloc[-1] - covid_data_2[covid_data_2['Country/Region'] == 'Vietnam']['confirmed'].iloc[-2]
value_confirmed
```

```python
covid_data_3 = covid_data_2[covid_data_2['Country/Region'] == 'Afghanistan'][['Country/Region', 'date', 'confirmed']].reset_index()
covid_data_3
```

```python
# 금일 확진자 수
covid_data_3['daily confirmed'] = covid_data_3['confirmed'] - covid_data_3['confirmed'].shift(1)
```

```python
# 최근 7일간의 평균 확진자수
covid_data_3['Rolling Ave.'] = covid_data_3['daily confirmed'].rolling(window=7).mean()
```
