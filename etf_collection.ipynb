## requests로 HTTP 요청

```python
import pandas as pd
import numpy as np
import requests
```

```python
url = 'https://finance.naver.com/api/sise/etfItemList.nhn?etfType=0&targetColumn=market_sum&sortOrder=desc'
```

```python
response = requests.get(url)
response
```

### json 타입으로 데이터 받기
json type으로 받아지 않을 경우에는
respnse.text가 json형식으로 이루어져 있는지 확인해보고,
그렇지 않다면 해결방법을 찾아볼 것

```python
etf_json = response.json()
etfItemList = etf_json["result"]["etfItemList"]
etfItemList
```

```python
len(etfItemList)
```

### json 데이터를 DataFrame으로 변환

```python
df = pd.DataFrame(etfItemList)
df.head()
```

```python
df.tail()
```

## 파일 저장

```python
# 날짜를 만들어 저장하기 위해 오늘 날짜를 구함
import datetime
today = datetime.datetime.today()
today = today.strftime("%Y-%m-%d")
today
```

```python
file_name = f"etf_{today}_raw.csv"
file_name
```

```python
df.to_csv(file_name, index=False)
```

```python
# itemcode가 int형으로 저장되었기 때문에 object형으로 변환시켜줘야 함
pd.read_csv(file_name, dtype={"itemcode": np.object})
```
