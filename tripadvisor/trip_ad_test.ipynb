```python

def move_page(page):
    
    page = page * 10
    
    url = f'https://www.tripadvisor.co.kr/Attraction_Review-g187147-d188150-Reviews-or{page}-Musee_d_Orsay-Paris_Ile_de_France.html'
    
    return url
```

```python

#    
for page in range(1, 44):
    
    url = move_page(page)
    
    driver.get(url)

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    avg_score_div = soup.select('div.WlYyy.cPsXC.fksET.cMKSg')

    # _list_original
    score_list = soup.select('svg.RWYkj.d.H0')

    index = 0
    #   
    score_array = []

    for score in score_list:
        if(22 <= index and 31 >= index):
                score_array.append(score.get('title')[8:])
        index = index + 1
        
    # 
    title_list = soup.select('span.NejBf')

    # 
    trip_date_list = soup.select('div.fEDvV')

    # 
    review_list = soup.select('div.WlYyy.diXIH.dDKKM > span.NejBf')
    
    for score, title, trip_date, review, in zip(score_array, title_list, trip_date_list, review_list):

        rows.append({

            'score' : score,
            'title' : title.get_text(),
            'trip_date' : trip_date.get_text(),
            'review' : review.get_text()
        })
        
        
```

```python
with open('trip_ad_test.csv', 'w', encoding='utf-8', newline="")as f:
    wr = csv.DictWriter(f, fieldnames = rows[0].keys())
    wr.writeheader()
    wr.writerows(rows)
```
